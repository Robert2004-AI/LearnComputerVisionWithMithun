# Task 1:

Perform standard PCA with 70% of the training dataset and identify the eigenvectors associated with top eigenvalues with 95% total energy. With these, train a logistic regression classifier to classify the images into 10 classes. Draw the ROC curve for the test dataset. Repeat the same with randomized PCA and compare.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.decomposition import TruncatedSVD
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, auc
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
import tensorflow as tf
from tensorflow.keras import layers, models
from sklearn.datasets import fetch_openml
from tensorflow.keras.utils import to_categorical

# ---- Load MNIST Dataset ----
def load_data():
    """
    Load the MNIST dataset from OpenML, normalize the pixel values to [0, 1],
    and split into training and testing sets.
    """
    mnist = fetch_openml('mnist_784', version=1)  # Fetch MNIST dataset
    X = mnist.data / 255.0  # Normalize pixel values to [0, 1]
    y = mnist.target.astype(int)  # Convert target labels to integers
    # Split the data into training (70%) and testing (30%) sets
    return train_test_split(X, y, test_size=0.3, random_state=42)

X_train, X_test, y_train, y_test = load_data()
```

```python
# Standard PCA
def perform_pca(X_train, X_test, energy=0.95):
    """
    Perform Principal Component Analysis (PCA) on the dataset.
    Reduce dimensionality while retaining the specified amount of explained variance (energy).

    Parameters:
        X_train (array): Training data.
        X_test (array): Test data.
        energy (float): Fraction of variance to preserve (default: 95%).

    Returns:
        PCA object, transformed training data, and transformed test data.
    """
    pca = PCA(n_components=energy)  # Initialize PCA to retain 'energy' variance
    X_train_pca = pca.fit_transform(X_train)  # Fit and transform training data
    X_test_pca = pca.transform(X_test)  # Transform test data
    return pca, X_train_pca, X_test_pca

# Randomized PCA (Truncated SVD)
def perform_randomized_pca(X_train, X_test, energy=0.95):
    """
    Perform dimensionality reduction using Truncated SVD (randomized PCA approximation).

    Parameters:
        X_train (array): Training data.
        X_test (array): Test data.
        energy (float): Approximation for the retained variance.

    Returns:
        TruncatedSVD object, transformed training data, and transformed test data.
    """
    svd = TruncatedSVD(n_components=int(X_train.shape[1] * energy))  # Use energy fraction of features
    X_train_svd = svd.fit_transform(X_train)  # Fit and transform training data
    X_test_svd = svd.transform(X_test)  # Transform test data
    return svd, X_train_svd, X_test_svd

# Logistic Regression and ROC Curve
def train_logistic_regression(X_train, X_test, y_train, y_test):
    """
    Train a logistic regression model on the training data and predict probabilities on the test set.

    Parameters:
        X_train (array): Transformed training data.
        X_test (array): Transformed test data.
        y_train (array): Training labels.
        y_test (array): Test labels.

    Returns:
        Predicted probabilities for the test set.
    """
    clf = LogisticRegression(max_iter=1000)  # Initialize logistic regression model
    clf.fit(X_train, y_train)  # Fit the model
    y_pred = clf.predict_proba(X_test)  # Predict class probabilities
    return y_pred

def plot_roc(y_test, y_pred, n_classes=10):
    """
    Plot ROC curves for each class using one-vs-all classification.

    Parameters:
        y_test (array): True test labels.
        y_pred (array): Predicted probabilities for the test set.
        n_classes (int): Number of classes (default: 10 for MNIST).
    """
    fpr, tpr, roc_auc = {}, {}, {}  # Initialize dictionaries to store metrics
    y_test_bin = to_categorical(y_test, num_classes=n_classes)  # One-hot encode test labels
    for i in range(n_classes):
        # Calculate FPR and TPR for each class
        fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_pred[:, i])
        # Compute AUC for each class
        roc_auc[i] = auc(fpr[i], tpr[i])
        # Plot ROC curve for the class
        plt.plot(fpr[i], tpr[i], label=f'Class {i} (AUC={roc_auc[i]:.2f})')
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.legend()  # Add legend
    plt.show()

# ---- Main Execution ----
# Task 1: Apply PCA and Evaluate Logistic Regression
print("Performing Standard PCA...")
pca, X_train_pca, X_test_pca = perform_pca(X_train, X_test)  # Perform PCA
y_pred_pca = train_logistic_regression(X_train_pca, X_test_pca, y_train, y_test)  # Train logistic regression
print("Plotting ROC for Standard PCA...")
plot_roc(y_test, y_pred_pca)  # Plot ROC curves

# Task 1: Apply Randomized PCA and Evaluate Logistic Regression
print("Performing Randomized PCA...")
svd, X_train_svd, X_test_svd = perform_randomized_pca(X_train, X_test)  # Perform randomized PCA
y_pred_svd = train_logistic_regression(X_train_svd, X_test_svd, y_train, y_test)  # Train logistic regression
print("Plotting ROC for Randomized PCA...")
plot_roc(y_test, y_pred_svd)  # Plot ROC curves
```
